```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/bone-marrow.csv')
```

```python
df.replace({'?': None},inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.drop('Disease',axis=1, inplace=True)
s = df.select_dtypes(include='object').columns
df[s] = df[s].astype("float")
```

```python
df.info()
```

```python
sns.set(rc = {'figure.figsize':(40,16)})
sns.heatmap(df.corr(),annot = True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ax = df['survival_status'].value_counts().plot(kind='bar')
ax.set_xticklabels(['dead','survived'])
ax.set(yLabel='Number of Patient')
plt.show()
```

```python
# Is donor age less than 35
ax = df['Donorage35'].value_counts().plot(kind='bar')
ax.set_xticklabels(['yes','no'])
ax.set(yLabel='Number of Patient')
plt.show()
```

```python
#Is recipient age below 10
ax = df['Recipientage10'].value_counts().plot(kind='bar')
ax.set_xticklabels(['yes','no'])
ax.set(yLabel='Number of Patient')
plt.show()
```
